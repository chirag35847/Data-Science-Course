## Maltplot example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup plot
fig,ax=plt.subplots(figsize=(5,5)) # (width,height)

# 3. Plot Data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",xlabel="x-axis",ylabel="y-axis")
```

## Two Options for subplots

```python
# Subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2, ncols=2, figsize=(20,10))
ax1.hist(x);
ax2.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax3.scatter(x,np.exp(x))
ax4.scatter(x,np.cos(x))
```

```python
# Option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(20,10))
ax[0,0].hist(x);
ax[0,1].scatter(x,np.sin(x))
ax[1,0].scatter(x,np.sin(x))
ax[1,1].bar(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
```

### Which one should we use
* When plotting quickly, pyplot is fine 
* When plotting something advanced use OO Method 

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",x='age',y='chol',c='target')
```

```python
fix,ax=plt.subplots(figsize=(10,6))
# ax.scatter(over_50["age"],over_50['chol'],c=over_50['target'])
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set_xlabel("AGE")
```

```python
fix,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50['chol'],c=over_50['target'])
ax.set(title="Heart Disease and Cholesterol Levels",xlabel="AGE",ylabel="Cholesterol Level")

# Add legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--")
```
